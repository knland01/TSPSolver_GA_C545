```python
import numpy as np

# Population of individuals
population = ["individual1", "individual2", "individual3", "individual4"]

# Corresponding fitness-based probabilities
fitness_scores = [10, 50, 30, 10]
total_fitness = sum(fitness_scores)
probabilities = [fitness / total_fitness for fitness in fitness_scores]

print(f"PROBABILITIES: {probabilities}")
# Select an individual with probabilities proportional to fitness
for i in range(30):
    selected_individual = np.random.choice(population, p=probabilities)
    print(selected_individual)


```
